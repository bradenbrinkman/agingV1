```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import data

# Excitatory spike counts
dataE = pd.read_csv('countSpikesExc.txt', sep="\t", header=None)
dataE.columns = ["30","35","40","45","50","55","60","65","70","75","80"]

dfE = pd.DataFrame(dataE) #convert data to a dataframe in pandas

# Inhibitory spike counts
dataI = pd.read_csv('countSpikesInh.txt', sep="\t", header=None)
dataI.columns = ["30","35","40","45","50","55","60","65","70","75","80"]

dfI = pd.DataFrame(dataI) #convert data to a dataframe in pandas

# Receptive field angles
dataR = pd.read_csv('RFangles.txt', sep="\t", header=None)
dataR.columns = ["30","35","40","45","50","55","60","65","70","75","80"]

dfR = pd.DataFrame(dataR) #convert data to a dataframe in pandas
```

```python
# Create colormap for violin plots
from matplotlib.cm import register_cmap
from matplotlib.colors import LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list(
    name='baroola', 
    colors=[sns.xkcd_rgb["dandelion"],sns.xkcd_rgb["dandelion"],sns.xkcd_rgb["goldenrod"],sns.xkcd_rgb["leaf green"],sns.xkcd_rgb["sea green"],sns.xkcd_rgb["aquamarine"],sns.xkcd_rgb["green blue"],sns.xkcd_rgb["cerulean blue"],sns.xkcd_rgb["deep sky blue"],sns.xkcd_rgb["electric blue"],sns.xkcd_rgb["pure blue"],sns.xkcd_rgb["true blue"],sns.xkcd_rgb["ultramarine"]]
)
register_cmap("mycolormap", cmap)
cpal = sns.color_palette("mycolormap", n_colors=11, desat=1.0)
```

```python
# Plot Excitatory spike counts
ax = sns.violinplot(data=dfE,scale='count',inner='box',cut=0,palette=cpal,linewidth=2); # cut=0 set the distribution range not exceeding the data range


plt.title("Excitatory spike counts",fontsize = 20);
plt.xlabel("Network age (training loops)",fontsize = 20);
plt.ylabel("Spikes fired",fontsize = 20);
plt.setp(ax.get_xticklabels(), fontsize=18);
plt.setp(ax.get_yticklabels(), fontsize=18);
plt.savefig("Ecounts_violin.pdf", bbox_inches='tight'); # save figure to pdf
```

```python
# Plot inhibitory spike counts
ax = sns.violinplot(data=dfI,scale='count',inner='box',cut=0,palette=cpal); # cut=0 set the distribution range not exceeding the data range

plt.title("Inhibitory spike counts",fontsize = 20);
plt.xlabel("Network age (training loops)",fontsize = 20);
plt.ylabel("Spikes fired",fontsize = 20);
plt.setp(ax.get_xticklabels(), fontsize=18);
plt.setp(ax.get_yticklabels(), fontsize=18);
plt.savefig("Icounts_violin.pdf", bbox_inches='tight'); # save figure to pdf
```

```python
# Plot receptive field angles
ax = sns.violinplot(data=dfR,scale='count',inner='box',cut=0,palette=cpal); # cut=0 set the distribution range not exceeding the data range

plt.title("Angle between receptive fields",fontsize = 20);
plt.xlabel("Network age (training loops)",fontsize = 20);
plt.ylabel("Angle (radians)",fontsize = 20);
plt.setp(ax.get_xticklabels(), fontsize=18);
plt.setp(ax.get_yticklabels(), fontsize=18);
plt.savefig("RFangleviolin.pdf", bbox_inches='tight'); # save figure to pdf
```
